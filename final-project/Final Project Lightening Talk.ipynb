## Function 1: Basic Species Data

### The purpose of these functions is to isolate data regarding strictly the types of species in the data set.
### A lot of the functions are continuations of each other, and piggy back off similar code, depending on what information I tried to get!


#### These first functions get the basic information I needed for my project

I gathered different information at first to try and see a general distribution of the species and what sort of data I was working with

```python
#list of species

in_file = open("pbdb_data_pilosa_full.csv", 'r', encoding= "ISO-8859-1") 
read_file = in_file.readlines()[18:] #starting where the occurance data starts

list_species = [] #empty list to store species in
       
for line in read_file:
    species = line.split('","')[11]
    name_of_species = line.split('","')[9]
    if "species" in species:  
        list_species.append(name_of_species)
print("\n".join(list_species)) #check to see if species list printed (works)
```

```python
#make a set to make a list of unique species
unique_species_set = set(list_species)
print("\n".join(unique_species_set))  #"\n".join prints list on new lines
```

```python
print(len(unique_species_set)) # to double check there were a large enough set of species for the project
```

```python
# creating a list of the mean ages of each species
mean_age_list = []


for line in read_file:
    minage = line.split('","')[15] #'""' gets rid of quotes
    maxage = line.split('","')[16]  
    mean_age = (float(minage)+float(maxage))/ 2 

    mean_age_list.append(float(mean_age))
print(mean_age_list) 
```

```python
#get an idea of the time frames of the data set with for loop

unique_means = set() #creating a set to get the unique mean age
for item in mean_age_list:
    unique_means.add(item)

unique_means_list = list(unique_means) #converting back to list
    
print(sorted(unique_means_list))  #sort numerically
```

```python
#finding which mean age had the most species present
from collections import Counter

ageDict = dict(Counter(mean_age_list))
print("{" + "\n".join("{}: {}".format(k, v) for k, v in ageDict.items()) + "}")

#0.06885: 127 most occurances late pleistocene
#1.29985: 110 second most pleistocene
```

## Basemap 
### library for plotting 2D data on maps

#### website with demo I followed for matplot: http://introtopython.org/visualization_earthquakes.html 

```python
from pylab import rcParams

%matplotlib inline
rcParams['figure.figsize'] = (8,6)

latitude = []
longitude = []

for line in read_file:
    lats = line.split('","')[30]
    lons = line.split('","')[29]
    latitude.append(float(lats))
    longitude.append(float(lons))  #redoing locational code from above to see it all together!!

# Make the plot larger.
plt.figure(figsize=(16,12))

    
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np

eq_map = Basemap(projection='robin', resolution = 'l', area_thresh = 1000.0,
                  lat_0=0, lon_0=-130)  #parameters for map
eq_map.drawcoastlines()
eq_map.drawcountries()
eq_map.fillcontinents(color = 'gray')
eq_map.drawmapboundary()
eq_map.drawmeridians(np.arange(0, 360, 30))
eq_map.drawparallels(np.arange(-90, 90, 30))

x,y = eq_map(longitude, latitude)
eq_map.plot(x, y, 'ro', markersize=6)


plt.figtext(.5,.9,'Stratigraphic Location: Pilosa', fontsize=40, ha='center')


plt.show()
```

## To try and find if any trends or patterns existed, I combined the different functions throughout my project now combining *age* with the species and locational data

```python
from pylab import rcParams

%matplotlib inline
rcParams['figure.figsize'] = (8,6)


latitude = []
longitude = []
# i have a list of means called mean_age_list already from above

for line in read_file:
    lats = line.split('","')[30]
    lons = line.split('","')[29]
    latitude.append(float(lats))
    longitude.append(float(lons))

# Make this plot larger.
plt.figure(figsize=(20,15))

    
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np


## used all the data collected from before about species and ages, to divide up the plot into three time scale categories

def get_marker_color(mean_age_list):
    if mean_age_list < 5.33:  #pliocene to present
        return ('go')
    elif mean_age_list < 23.03:  #miocene
        return ('yo')
    else:
        return ('ro')  #oligocene to the beginning of Cenozoic (all of this is in Cenozoic)


eq_map = Basemap(projection='robin', resolution = 'l', area_thresh = 1000.0,
                  lat_0=0, lon_0=-130)
eq_map.drawcoastlines()
eq_map.drawcountries()
eq_map.fillcontinents(color = 'gray')
eq_map.drawmapboundary()
eq_map.drawmeridians(np.arange(0, 360, 30))
eq_map.drawparallels(np.arange(-90, 90, 30))

for lon, lat, age in zip(longitude, latitude, mean_age_list):
    
    x,y = eq_map(lon, lat)
    marker_string = get_marker_color(age)
    eq_map.plot(x, y, marker_string, markersize=8)

plt.figtext(.5,.9,'Stratigraphic Location: Pilosa 2', fontsize=30, ha='center')

plt.show()
```
