```python
import csv
from pylab import rcParams

%matplotlib inline
rcParams['figure.figsize'] = (8,6)

filename = 'formatted_sloth_pbdb_final_data.csv'

lats, lons = [], []

with open (filename) as f:
    reader = csv.reader(f)
    
    for i in range(19):
        reader.__next__()
    
    for row in reader:
        lats.append(float(row[29]))
        lons.append(float(row[30]))
        
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
 
eq_map = Basemap(projection='mill', lat_0=50, lon_0=-100,
resolution='c', area_thresh=1000.0)

eq_map.drawcoastlines()
eq_map.drawcountries()
eq_map.fillcontinents(color='white', lake_color='aqua')
eq_map.drawmapboundary(fill_color='lightblue')

x, y = eq_map(lons, lats)
eq_map.plot(x, y, 'ro', markersize=4)

plt.show()
```
